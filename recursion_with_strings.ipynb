```python
# Palindrome Check using Recursion
# Problem Description:

# You are given a string. Your task is to check if the string is a palindrome using recursion. A palindrome is a string that reads the same forward and backward.



# Input:

# A string s.



# Output:

# A boolean indicating whether the string is a palindrome.



# Example:

# Input: s = "radar"
# Output: True
 
# Input: s = "hello"
# Output: False
```

```python
def is_palindrome(s):
    """
    Function to check if a string is a palindrome using recursion.
    
    Parameters:
    s (str): The string to check.
    
    Returns:
    bool: True if the string is a palindrome, False otherwise.
    """
    def is_palindrome_helper(s,start,end):
        if start>=end:
            return True
        if s[start]!=s[end]:
            return False 
        return is_palindrome_helper(s,start+1,end-1)
    return is_palindrome_helper(s,0,len(s)-1)
```

```python
# Subsequences of a String
# Problem Description:

# You are given a string. Your task is to return all possible subsequences of that string using recursion. A subsequence is a sequence that can be derived from another sequence by deleting some elements without changing the order of the remaining elements.



# Input:

# A string s.



# Output:

# A list of all possible subsequences of the string.



# Example:

# Input: s = "abc"
# Output: ['', 'c', 'b', 'bc', 'a', 'ac', 'ab', 'abc']
 
# Input: s = "a"
# Output: ['', 'a']
```

```python
def subsequences(s):
    """
    Function to generate all subsequences of a string using recursion.
    
    Parameters:
    s (str): The string to generate subsequences from.
    
    Returns:
    list of str: A list containing all subsequences of the string.
    """
    if s=='':
        return ['']
    smallAnswer=subsequences(s[1:])
    ans=[]
    ans.extend(smallAnswer)
    for subsequence in smallAnswer:
        ans.append(s[0]+subsequence)
    
    return ans
```

```python
# Permutation of a String
# Problem Description:

# You are given a string. Your task is to return all possible permutations of that string using recursion. A permutation of a string is a rearrangement of its characters in all possible ways.



# Input:

# A string s.



# Output:

# A list of all possible permutations of the string.



# Example:

# Input: s = "abc"
# Output: ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']
 
# Input: s = "a"
# Output: ['a']
```

```python
def permutations(s):
    """
    Function to generate all permutations of a string using recursion.
    
    Parameters:
    s (str): The string to generate permutations from.
    
    Returns:
    list of str: A list containing all permutations of the string.
    """
    if s=='':
        return ['']
    c1=s[0]
    sub_permutations=permutations(s[1:])
    ans=[]
    for eachPerm in sub_permutations:
        for pos in range(0,len(eachPerm)+1):
            ans.append(eachPerm[0:pos]+c1+eachPerm[pos:])
    return ans
```
