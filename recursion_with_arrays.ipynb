```python
# Sum of an Array using Recursion
# Problem Description:

# You are given an array of integers. Your task is to calculate and return the sum of all the elements in the array using recursion.



# Input:

# An array of integers.



# Output:

# An integer representing the sum of the array elements.



# Example:

# Input: arr = [1, 2, 3, 4, 5]
# Output: 15
 
# Input: arr = [10, -2, 3, 5]
# Output: 16
```

```python
def recursive_sum(arr):
    """
    Function to calculate the sum of an array using recursion.
    
    Parameters:
    arr (list of int): The array of integers.
    
    Returns:
    int: The sum of the array elements.
    """
    # Your code here
    if len(arr)==0:
        return 0 
    sum=arr[0]+recursive_sum(arr[1:])
    return sum
```

```python
# First Index of an Element using Recursion
# Problem Description:

# You are given an array and an element. Your task is to find the first index where the array contains that element using recursion. If the element is not found, return -1.



# Input:

# An array arr of integers.

# An integer element.



# Output:

# An integer representing the first index where element is found in arr, or -1 if it is not found.



# Example:

# Input: arr = [1, 2, 3, 2, 4, 2], element = 2
# Output: 1
 
# Input: arr = [5, 6, 7, 8], element = 10
# Output: -1
```

```python
def find_first_index(arr, element):
    """
    Function to find the first index of a given element in an array using recursion.
    
    Parameters:
    arr (list of int): The array to search through.
    element (int): The element to find.
    
    Returns:
    int: The first index of the element in the array, or -1 if not found.
    """
    # Your code here
    if len(arr)==0:
        return -1
    if arr[0]==element:
        return 0
    ans=find_first_index(arr[1:],element)
    if ans==-1:
        return ans 
    else:
        return ans+1
    
```

```python
# Print all Index of Element using Recursion
# Problem Description:

# You are given an array and an element. Your task is to find all the indices where the array contains that element using recursion.



# Input:

# An array arr of integers.

# An integer element.



# Output:

# A list of indices where element is found in arr.



# Example:

# Input: arr = [1, 2, 3, 2, 4, 2], element = 2
# Output: [1, 3, 5]
 
# Input: arr = [5, 6, 7, 8], element = 10
# Output: []
```

```python
def find_indices(arr, element):
    """
    Function to find all indices of a given element in an array using recursion.
    
    Parameters:
    arr (list of int): The array to search through.
    element (int): The element to find.
    
    Returns:
    list of int: A list containing all indices of the element in the array.
    """
    # Your code here
    lst=[]
    def find_indices_helper(arr,element,index):
        if len(arr)==index:
            return 
        if arr[index]==element:
            lst.append(index)
        find_indices_helper(arr,element,index+1)
    find_indices_helper(arr,element,0)
    return lst
```
